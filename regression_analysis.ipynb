```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("final_ess.csv")
df.head()
```

```python
# Variables to compare with climate_worry
variables = [
    "age",
    "income",
    "right_orientation",
    "happiness",
    "life_satisfaction",
    "expo",
    "years_education"
]

for var in variables:
    plt.figure(figsize=(6,4))

    x_jitter = df[var] + np.random.uniform(-0.15, 0.15, size=len(df))
    y_jitter = df["climate_worry"] + np.random.uniform(-0.15, 0.15, size=len(df))

    plt.scatter(x_jitter, y_jitter, alpha=0.4, s=20)

    plt.xlabel(var)
    plt.ylabel("climate_worry")
    plt.title(f"Climate Worry vs {var}")

    plt.show()
```
